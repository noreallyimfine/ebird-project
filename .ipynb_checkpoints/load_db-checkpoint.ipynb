```python
from ebird import db
```

```python
from ebird.models import Bird, County, State
```

```python
import pandas as pd
birds = pd.read_csv('C:/Users/jayja/Desktop/bird.csv')
```

```python
for bird in birds['0']:
    input = Bird(name=bird)
    db.session.add(input)

db.session.commit()
```

```python
from joblib import load
```

```python
county_state = load('C:/Users/jayja/Desktop/old-ebird/bird_app/utils/county_state.joblib')
```

```python
county_state = [cs.split(',') for cs in county_state]
county_state
```

```python
len(county_state)
```

```python
county_state = sorted(county_state, key=lambda x: (x[1], x[0]))
county_state
```

```python
county_state[0]
```

```python
states = sorted(list(set([cs[1] for cs in county_state])))
states
```

```python
len(states)
```

```python
for state in states:
    state_in = State(name=state)
    db.session.add(state_in)
    
db.session.commit()
```

```python
for cs in county_state:
    county_in = County(county_name=cs[0], state=cs[1])
    db.session.add(county_in)

db.session.commit()
    
```
