```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\ajaco\\Desktop\\repos\\noreallyimfine\\ebird-project\\data\\ebd_relJan-2020.txt', sep='\t', nrows=200000, usecols=['COMMON NAME', 'COUNTRY', 'STATE', 'COUNTY', 'LATITUDE', 'LONGITUDE', 'OBSERVATION DATE', 'OBSERVATION COUNT'])

print(df.shape)

df.head()
```

```python
us_birds = df[df['COUNTRY'] == 'United States']

print(us_birds.shape)
```

```python
us_birds['OBSERVATION DATE'] = pd.to_datetime(us_birds['OBSERVATION DATE'], infer_datetime_format=True)
```

```python
us_birds['observ_year'] = us_birds['OBSERVATION DATE'].dt.year
us_birds['observ_month'] = us_birds['OBSERVATION DATE'].dt.month
```

```python
print(us_birds['observ_year'].min())
print(us_birds['observ_year'].max())
```

```python
us_birds[us_birds['observ_year'] < 1900]
```

```python
us_birds[us_birds['observ_year'] > 1950]
```

```python
def month_to_season(x):
    if x in [12, 1, 2]:
        return 'Winter'
    elif x in [3, 4, 5]:
        return 'Spring'
    elif x in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

us_birds['season'] = us_birds['observ_month'].apply(month_to_season)
```

```python
us_birds.head()
```

```python
us_birds['OBSERVATION COUNT'] = us_birds['OBSERVATION COUNT'].map(lambda x: x if x != 'X' else 1)
```

```python
us_birds.rename(columns={'COMMON NAME': 'common_name',
                         'OBSERVATION COUNT': 'observ_count',
                         'COUNTRY': 'country',
                         'STATE': 'state',
                         'COUNTY': 'county',
                         'LATITUDE': 'lat',
                         'LONGITUDE': 'lon',
                         'OBSERVATION DATE': 'observ_date'}, inplace=True)

us_birds.head()
```

```python
# save birds file so we dont have to rebuild in the future
us_birds.to_csv('us_birds.csv', index=False)
```
