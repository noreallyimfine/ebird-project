```python
df = pd.read_csv("us_birds.csv")

print(df.shape)

df.head()
```

```python
df.rename(columns={
    'common_name': 'name',
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['county'], inplace=True)
```

```python
df['bad_name'] = df['name'].apply(lambda x: 0 if ("sp." in x) or ("(" in x) or ("/" in x) else 1)
```

```python
df = df.drop(columns=['bad_name'])
```

```python
df['observ_count'].value_counts()
```

```python
df.dtypes
```

```python
df['observ_date'] = pd.to_datetime(df['observ_date'], infer_datetime_format=True)
```

```python
df.dtypes
```

```python
assert('North Slope Borough' not in df['county'].values)
```

```python
assert('West Baton Rouge' in df['county'].values)
```

```python
season_region_ct = pd.pivot_table(df,
                                  index='name',
                                  columns=['region', 'season'],
                                  values='observ_count',
                                  aggfunc='count',
                                  fill_value=0.0)


joblib.dump(season_region_ct, 'lookup_table.joblib')
```
